```python
import pandas as pd
import numpy as np
```

```python
path = 'https://raw.githubusercontent.com/KyleYounghans/nflscrapR-data/master/games_data/regular_season/reg_games_2018.csv'
games2018b = pd.read_csv(path)
games2018dfb = pd.DataFrame(games2018b)
winLoss = games2018dfb[['home_team', 'away_team', 'home_score', 'away_score']].copy()
winLoss['homeWin'] = None
```

```python
winLoss.head()
```

```python
winLoss.loc[winLoss['home_score'] > winLoss['away_score'], 'homeWin'] = 1
#homeWin = home_team losing at home = 0
winLoss.loc[winLoss['home_score'] < winLoss['away_score'], 'homeWin'] = 0
#homeWin = home_team tieing at home = 0.5
winLoss.loc[winLoss['home_score'] == winLoss['away_score'], 'homeWin'] = 0.5
```
