```python
import pandas as pd
```

```python
data2009 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2009.csv"
data2010 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2010.csv"
data2011 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2011.csv"
data2012 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2012.csv"
data2013 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2013.csv"
data2014 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2014.csv"
data2015 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2015.csv"
data2016 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2016.csv"
data2017 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2017.csv"
data2018 = "https://raw.githubusercontent.com/ryurko/nflscrapR-data/master/games_data/regular_season/reg_games_2018.csv"
```

```python
urls = [data2009, data2010, data2011, data2012, data2013, data2014, data2015, data2016, data2017, data2018]
```

```python
combined_csv = pd.concat([pd.read_csv(url) for url in urls])
```

```python
combined_csv.head()
```

```python
#clean csv files
cleanDF2 = combined_csv.drop(columns=['type','game_id','week','state_of_game','game_url'])
cleanDF2.head()
```

```python
#generate a variable counting wins and losses for home and away teams

#homeWin = home_team winning at home = 1
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'homeWin'] = 1
#homeWin = home_team losing at home = 0
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'homeWin'] = 0
#homeWin = home_team tieing at home = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'homeWin'] = 0.5
#homeLoss = home_team losing at home = 1
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'homeLoss'] = 1
#homeLoss = home_team winning at home = 0
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'homeLoss'] = 0
#homeLoss = home_team tieing at home = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'homeLoss'] = 0.5
#awayWin = away_team losing away = 0
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'awayWin'] = 0
#awayWin = away_team winning away = 1
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'awayWin'] = 1
#awayWin = away_team tieing away = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'awayWin'] = 0.5
#awayloss = away_team winning away = 0
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'awayLoss'] = 0
#awayloss = away_team losing away = 1
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'awayLoss'] = 1
#awayloss = away_team tieing away = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'awayLoss'] = 0.5
cleanDF2.head()
```

```python


cleanDF2['homeWin'] = cleanDF2['homeWin'].astype(int)
cleanDF2['homeLoss'] = cleanDF2['homeLoss'].astype(int)
cleanDF2['awayLoss'] = cleanDF2['awayLoss'].astype(int)
cleanDF2['awayWin'] = cleanDF2['awayWin'].astype(int)
cleanDF2.head()

```

```python
#home team 
cDF2 = cleanDF2.drop(columns=['away_team','awayWin','awayLoss'])
cDF2
```

```python
#home team
TableHT = cDF2.rename(columns={"home_team": "Team", "home_score": "own_score@home", "away_score": "other_score@home"})
TableHT.head()
```

```python
#Home team
HTs = TableHT.groupby(['season']).sum()
HTs
```

```python
#home team
HT = TableHT.groupby(['season']).mean()
HT.head()
```

```python
#away team
cDF3 = cleanDF2.drop(columns=['home_team','homeWin','homeLoss'])
cDF3.head()

```

```python
#Away Team
TableAT = cDF3.rename(columns={"away_team": "Team", "home_score": "other_score@away", "away_score": "own_score@away"})
TableAT.head()
```

```python
#Away Team
ATs = TableAT.groupby(['season']).sum()
ATs.head()
```

```python
AT = TableAT.groupby(['season']).mean()
AT.head()
```

```python
#combined table summing all the data
combined_table_S=pd.merge(ATs, HTs, on=["season"])
combined_table_S.head()
```

```python
#combined table averaging all the data
combined_table=pd.merge(AT, HT, on=["season"])
combined_table.head()

```

```python
combined_table['Win Avg'] = (combined_table_S['awayWin'] + combined_table_S['homeWin'])/(combined_table_S['awayWin'] + combined_table_S['homeWin'] + combined_table_S['awayLoss'] + combined_table_S['homeLoss'] )
combined_table.head()
```

```python
CT = combined_table.rename(columns={"other_score@away": "AwayScore",})
CT.head()
```

```python
Summary_T = CT[['homeWin','awayWin','own_score@home','other_score@home']].round(2)
Summary_T
```

```python
Summary_Table = Summary_T.rename(columns={"homeWin": "HomeWin", "awayWin": "AwayWin", "own_score@home": "HomeScore", "other_score@home": "AwayScore"})
Summary_Table.head()
```

```python
overall_average = Summary_Table.append(Summary_Table.agg(['mean'])).round(2)
overall_average
```

```python
overall_average.to_json(r'Json files\Overall_Average_Table.json')
overall_average.to_csv(r'CSV files\Overall_Average_Table.csv')
```
