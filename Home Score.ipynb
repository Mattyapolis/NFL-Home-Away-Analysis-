```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/KyleYounghans/nflscrapR-data/master/games_data/regular_season/reg_games_2018.csv'
games2018b = pd.read_csv(path)
games2018dfb = pd.DataFrame(games2018b)
cleanDF2 = games2018dfb[['home_team', 'away_team', 'home_score', 'away_score']].copy()
cleanDF2['homeWin'] = None
```

```python
cleanDF2.head()

```

```python
#generate a variable counting wins and losses for home and away teams

#homeWin = home_team winning at home = 1
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'homeWin'] = 1
#homeWin = home_team losing at home = 0
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'homeWin'] = 0
#homeWin = home_team tieing at home = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'homeWin'] = 0.5
#homeLoss = home_team losing at home = 1
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'homeLoss'] = 1
#homeLoss = home_team winning at home = 0
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'homeLoss'] = 0
#homeLoss = home_team tieing at home = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'homeLoss'] = 0.5
#awayWin = away_team losing away = 0
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'awayWin'] = 0
#awayWin = away_team winning away = 1
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'awayWin'] = 1
#awayWin = away_team tieing away = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'awayWin'] = 0.5
#awayloss = away_team winning away = 0
cleanDF2.loc[cleanDF2['home_score'] < cleanDF2['away_score'], 'awayLoss'] = 0
#awayloss = away_team losing away = 1
cleanDF2.loc[cleanDF2['home_score'] > cleanDF2['away_score'], 'awayLoss'] = 1
#awayloss = away_team tieing away = 0.5
cleanDF2.loc[cleanDF2['home_score'] == cleanDF2['away_score'], 'awayLoss'] = 0.5
cleanDF2.head()
```

```python


cleanDF2['homeWin'] = cleanDF2['homeWin'].astype(int)
cleanDF2['homeLoss'] = cleanDF2['homeLoss'].astype(int)
cleanDF2['awayLoss'] = cleanDF2['awayLoss'].astype(int)
cleanDF2['awayWin'] = cleanDF2['awayWin'].astype(int)
cleanDF2.head()

```

```python
#home team 
cDF2 = cleanDF2.drop(columns=['away_team','awayWin','awayLoss'])
cDF2.head()
```

```python
#home team
TableHT = cDF2.rename(columns={"home_team": "Team", "home_score": "own_score@home", "away_score": "other_score@home"})
TableHT.head()
```

```python
#Home team
HTs = TableHT.groupby('Team').sum()
HTs
```

```python
#home team
HT = TableHT.groupby('Team').mean()
HT
```

```python
#away team
cDF3 = cleanDF2.drop(columns=['home_team','homeWin','homeLoss'])
cDF3.head()

```

```python
#Away Team
TableAT = cDF3.rename(columns={"away_team": "Team", "home_score": "other_score@away", "away_score": "own_score@away"})
TableAT.head()
```

```python
#Away Team
ATs = TableAT.groupby('Team').sum()
ATs.head()
```

```python
AT = TableAT.groupby('Team').mean()
AT.head()
```

```python
#combined table summing all the data
combined_table_S=pd.merge(ATs, HTs, on=["Team"])
combined_table_S.head()
```

```python
#combined table averaging all the data
combined_table=pd.merge(AT, HT, on=["Team"])
combined_table.head()

```

```python
combined_table['Win %'] = (combined_table_S['awayWin'] + combined_table_S['homeWin'])/16
combined_table.head()
```

```python
CT = combined_table.rename(columns={"awayWin": "awayWin %", "awayLoss": "awayLoss %", "homeWin": "homeWin %", "homeLoss": "homeLoss %"})
CT.head()
```

```python
Summary_Table = CT[['Win %', 'homeWin %','homeLoss %', 'awayWin %', 'awayLoss %', 'own_score@home','other_score@home','own_score@away','other_score@away']]
Summary_Table.head()
```

```python
Summary_Table.to_json(r'Json files\Home_win_score.json')
Summary_Table.to_csv(r'CSV files\Home_win_score.csv')
```
