```python
import pandas as pd
```

```python
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 

stadium_data = pd.read_csv("nfl_stadiums.csv") 

# Preview the first 5 lines of the loaded data 
stadium_data.head()
```

```python
stadium_data_C = stadium_data.rename(columns={"team_abbv": "Team"})
stadium_data_C.head()
```

```python
Coordinates = stadium_data_C.drop(columns=['stadium_home_team','stadium_name','stadium_name','stadium_surface','stadium_roof_type', 'stadium_capacity','stadium_location','stadium_opened'])
Coordinates.head()
```

```python
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 

team_perf = pd.read_csv('CSV files/Home_win_score.csv') 

# Preview the first 5 lines of the loaded data 
team_perf.head()
```

```python
#combined table averaging all the data
combined_table_C=pd.merge(team_perf, Coordinates, on=["Team"])
combined_table_C.head()
```

```python
combined_table_C.to_json(r'Json files\Home_win_score_Coordinates.json')
combined_table_C.to_csv(r'CSV files\Home_win_score_Coordinates.csv')
```

```python
#combined table averaging all the data
combined_table=pd.merge(team_perf, stadium_data_C, on=["Team"])
combined_table.head()
```

```python
by_surface = combined_table.drop(columns=['Team','stadium_home_team','stadium_name','stadium_name','stadium_capacity','stadium_location','stadium_opened','stadium_latitude','stadium_longitude'])
by_surface.head()
```

```python
surface = by_surface.groupby('stadium_surface').mean()
surface.head()
```
