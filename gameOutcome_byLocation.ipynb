```python
# Import dependencies
import pandas as pd

#Include all SportsReference for now
from sportsreference.nfl.boxscore import Boxscore
from sportsreference.nfl.schedule import Schedule
from sportsreference.nfl.teams import Teams
```

```python
# Get abbreviations
abbreviations = Teams()
for team in abbreviations:
    print(team.name, team.abbreviation)  # Prints the team's name and abbreviation
```

```python
# Make into a list
# Get all team abbreviations in an array
abbrList = []
for team in abbreviations: 
    abbrList.append(team.abbreviation)
print(abbrList)
```

```python
# Loop attempt one. (Getting info as list)
for abbr in abbrList: 
    print(abbr)
    team_schedule = Schedule(abbr)
    print(f"Number games: {len(team_schedule)}")
    for game in team_schedule:
        print(game.location, game.result)
        
```

```python
# Refined loop, get ID and percentage of home wins
for abbr in abbrList: 
    print(abbr)
    team_schedule = Schedule(abbr)
    homeCount = 0
    homeWin = 0
    awayCount = 0
    awayWin = 0
    for game in team_schedule:
        if game.location == "Home":
            homeCount += 1
            if game.result == "Win":
                homeWin += 1 
        elif game.location =="Away":
            awayCount += 1
            if game.result =="Win":
                awayWin += 1
    print(f"Home win rate: {(homeWin/homeCount)*100}%")
    print(f"Away win rate: {(awayWin/awayCount)*100}%")
    

    
```

```python
## Next step: make this into a chart/dataframe? JSON file? How do we want to export this? 
outcomeData = []

for abbr in abbrList: 
    print(f"Checking {abbr} data")
    team_schedule = Schedule(abbr)
    homeCount = 0
    homeWin = 0
    awayCount = 0
    awayWin = 0
    for game in team_schedule:
        if game.location == "Home":
            homeCount += 1
            if game.result == "Win":
                homeWin += 1 
        elif game.location =="Away":
            awayCount += 1
            if game.result =="Win":
                awayWin += 1
    homeWinRate = homeWin/homeCount
    awayWinRate = awayWin/awayCount
    outcomeData.append({"team":abbr, "homeWinRate": homeWinRate, "awayWinRate": awayWinRate})
    

    
```

```python
outcomeDataFrame = pd.DataFrame(outcomeData)
```
