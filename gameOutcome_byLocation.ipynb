```python
# Import dependencies
import pandas as pd

#Include all SportsReference for now
from sportsreference.nfl.boxscore import Boxscore
from sportsreference.nfl.schedule import Schedule
from sportsreference.nfl.teams import Teams
```

```python
# Get abbreviations
abbreviations = Teams()
for team in abbreviations:
    print(team.name, team.abbreviation)  # Prints the team's name and abbreviation
```

```python
# Make into a list
# Get all team abbreviations in an array
abbrList = []
for team in abbreviations: 
    abbrList.append(team.abbreviation)
print(abbrList)
```

```python
# Loop attempt one. 
for abbr in abbrList: 
    print(abbr)
    team_schedule = Schedule(abbr)
    for game in team_schedule:
        print(game.location, game.result)
```
